```python
# Recreating Reverse version of Range() using iterator
class ReverseRange:
    def __init__(self,sta,sto=0,ste=1):
        self.start=sta
        self.step=ste
        self.stop=sto
    def __iter__(self):
        return self
    def __next__(self):
        if (self.start<self.stop):
            raise StopIteration
        value =self.start
        self.start=self.start-self.step
        return value

test=ReverseRange(1,10,-2)
for i in test:
    print(i)
```

```python
#generating primes upto n using generators
def prime_generator(n):
    for i in range(2,n):
        k=0
        for j in range(2,int(i**0.5)+1):
            if (i%j==0):
                k=1
                break
        if (k==0):
            yield i
        else :
            continue
X=int(input("Enter a number : "))
for i in prime_generator(X):
    print(i)
```

```python
#generator using next
def prime():
    i=2
    while True:
        k=0
        for j in range (2,int(i**0.5)+1):
            if(i%j==0):
                k=1
                break
        if (k==0):
            yield i
        i+=1
g=prime()
print(next(g))
print(next(g))
print(next(g))




```

```python
import dataclasses
#Encapsulation + decorator
class BankAccount:
    def __init__(self):
        self.__balance=0
    @property
    def balance(self):
        return(self.__balance)
    @balance.setter
    def balance(self,amount):
        if(amount+self.__balance<0):
            raise ValueError
        self.__balance=self.__balance+amount
        if(amount<=0):
            print(f"{amount} withdrawn from account. Current balance : {self.__balance}")
        else:
            print(f"{amount} deposited into account. Current balance : {self.__balance}")
A=BankAccount()
A.balance=100
A.balance=-70
print(f"Current balance :{A.balance}")
```

```python
#polymorphism
class Employee:
    def __init__(self):
        self.name=""
        self.sal=0
    def calculate_payroll():
        pass
class Salaried(Employee):
    def __init__(self,name,salary):
        self.sal=salary
        self.name=name
    def calculate_payroll(self):
        print(f"Yearly salary of {self.name} is {(self.sal*12)}")
class Hourly(Employee):
    def __init__(self,name,hsalary,hours):
        self.sal=hsalary
        self.hours=hours
        self.name=name
    def calculate_payroll(self):
        print(f"Salary of {self.name} is {(self.sal*self.hours)}")

def payroll(type):
    return(type.calculate_payroll())
l=[Salaried("Alex",1000), Hourly("Alan", 10, 30)]
for i in l:
    payroll(i)
```

```python
class A:
    def process(self): print("A")

class B(A):
    def process(self): print("B"); super().process()

class C(A):
    def process(self): print("C"); super().process()

class D(B, C):
    def process(self): print("D"); super().process()

obj=D()
obj.process()
```

```python
#Duck Typing
class email:
    def broadcast(self,message):
        print(f"\"{message}\" sent via email")
class sms:
    def broadcast(self,message):
        print(f"\"{message}\" sent via sms")
class slack:
    def broadcast(self,message):
        print(f"\"{message}\" sent via slack")
def send_broadcast(channel,message):
    channel.broadcast(message)
send_broadcast(sms(),"Hi there")
send_broadcast(email(),"How are you doing ?")
```

```python
def calculate_area(*args):
    if(len(args)==1):
        print(f"Area of circle is {3.14*args[0]**2}")
    else:
        print(f"Area of rectangle is {args[0]*args[1]}")

calculate_area(4,5)
calculate_area(9)
```
